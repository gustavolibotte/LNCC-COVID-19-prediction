```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x, y, z, t, x1, x2, y1, y2, z1, z2, k, omega, E = sp.symbols("x y z t x_1 x_2 y_1 y_2 z_1 z_2 k omega E", real=True)
```

```python
E1 = E*sp.exp(sp.I*(k*sp.sqrt((x-x1)**2+(y-y1)**2+(z-z1)**2)-omega*t))/sp.sqrt((x-x1)**2+(y-y1)**2+(z-z1)**2)
E2 = E*sp.exp(sp.I*(k*sp.sqrt((x-x2)**2+(y-y2)**2+(z-z2)**2)-omega*t))/sp.sqrt((x-x2)**2+(y-y2)**2+(z-z2)**2)

# E1 = E*sp.exp(sp.I*(k*sp.sqrt(x**2+y**2+(z-z1)**2)-omega*t))/sp.sqrt(x**2+y**2+(z-z1)**2)
# E2 = E*sp.exp(sp.I*(k*sp.sqrt(x**2+y**2+(z-z2)**2)-omega*t))/sp.sqrt(x**2+y**2+(z-z2)**2)

sp.simplify(E1+E2)
```

```python
I = (E1+E2)*sp.conjugate(E1+E2)
I = sp.simplify(sp.expand(I))
# I = I.subs(x1, 0).subs(x2, 0).subs(y1, 0).subs(y2, 0)
I
```

```python
f = sp.lambdify((x, y, z, x1, x2, y1, y2, z1, z2, k, E), I)
```

```python
xl = np.linspace(-1, 1, 512)
yl = np.linspace(-1, 1, 512)
x_, y_ = np.meshgrid(xl, yl, indexing="ij")

plt.figure(figsize=(7, 5))
plt.contourf(xl, yl, f(x_, y_, 0, 0, 0, 0, 0, 1, 1, 100, 1).real.T, 100, cmap="plasma")
plt.colorbar()
plt.show()
```
