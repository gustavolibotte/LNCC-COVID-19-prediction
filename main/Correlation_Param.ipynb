# Correlation between parameters from SEIRD model

```
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
data = pd.read_csv('https://raw.githubusercontent.com/gustavolibotte/LNCC-COVID-19-prediction/master/EAMC_2021_paper/main/CorrelParams.csv')[:20]
data2 = pd.read_csv('https://raw.githubusercontent.com/gustavolibotte/LNCC-COVID-19-prediction/master/EAMC_2021_paper/main/CorrelParams.csv')[20:40]
data3 = pd.read_csv('https://raw.githubusercontent.com/gustavolibotte/LNCC-COVID-19-prediction/master/EAMC_2021_paper/main/CorrelParams.csv')[40:60]
data4 = pd.read_csv('https://raw.githubusercontent.com/gustavolibotte/LNCC-COVID-19-prediction/master/EAMC_2021_paper/main/CorrelParams.csv')[60:80]
data5 = pd.read_csv('https://raw.githubusercontent.com/gustavolibotte/LNCC-COVID-19-prediction/master/EAMC_2021_paper/main/CorrelParams.csv')[80:100]
```

```
corr = data.corr()
corr_2 = data2.corr()
corr_3 = data3.corr()
corr_4 = data4.corr()
corr_5 = data5.corr()
```

```
corr = np.array(corr)
corr_2 = np.array(corr_2)
corr_3 = np.array(corr_3)
corr_4 = np.array(corr_4)
corr_5 = np.array(corr_5)
```

```
total = (corr+corr_2+corr_3+corr_4+corr_5)/5
```

```
std = np.std([corr[0,1],corr_2[0,1],corr_3[0,1],corr_4[0,1],corr_5[0,1]])
```

```
std = []
for i in range(7):
  for j in range(7):
    std_new = np.std([corr[i,j],corr_2[i,j],corr_3[i,j],corr_4[i,j],corr_5[i,j]])
    std.append(std_new)
```

```
len(std)
```

```
ticks = [r'$\beta_I$',r'$\beta_E$', 'N', r'$P_{IFR}$', r'$\tau_d$', r'$\tau_r$', r'$\tau_c$']
```
